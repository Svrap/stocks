```python
import pandas as pd
pd.set_option('display.max_columns', None)

axis_bank_df = pd.read_csv("AXISBANK.csv")
f_axis_bank_df = axis_bank_df[['Date','Symbol','Open','Close','Volume']]
f_axis_bank_df.rename(columns={"Open": "Open_axis","Close": "Close_axis"},inplace= True)
```

```python
f_axis_bank_df_cp = f_axis_bank_df.copy()
f_axis_bank_df_cp['daily_value_axis']= f_axis_bank_df['Close_axis'] * f_axis_bank_df['Volume']
# f_axis_bank_df['daily_value'].astype(int)
f_axis_bank_df_cp
```

```python
adani_port_df = pd.read_csv("ADANIPORTS.csv")
f_adani_port_df = adani_port_df[['Date','Symbol','Open','Close','Volume']]
f_adani_port_df.rename(columns={"Open": "Open_adani","Close": "Close_adani"},inplace= True)
f_adani_port_df_cp = f_adani_port_df.copy()
f_adani_port_df_cp.loc[:,'daily_value_adani']  = f_adani_port_df['Close_adani'] * f_adani_port_df['Volume']
f_adani_port_df_cp
```

```python
asian_paints_df = pd.read_csv("ASIANPAINT.csv")
f_asian_paints_df = asian_paints_df[['Date','Symbol','Open','Close','Volume']]
f_asian_paints_df.rename(columns={"Open": "Open_asian","Close": "Close_asian"},inplace= True)
f_asian_paints_df_cp = f_asian_paints_df.copy()
f_asian_paints_df_cp.loc[:,'daily_value_asian_paint']  = f_asian_paints_df['Close_asian'] * f_asian_paints_df['Volume']
f_asian_paints_df_cp
```

```python
df = pd.concat([f_axis_bank_df_cp.set_index('Date'),f_adani_port_df_cp.set_index('Date'),f_asian_paints_df_cp.set_index('Date')], axis=1,join='inner').reset_index()
df
```

# EquityCurve

```python
df['equity_curve']=df[['daily_value_axis','daily_value_adani','daily_value_asian_paint']].sum(axis=1)
pd.set_option('display.float_format','{:.2f}'.format)
```

```python
df_cp = df.copy()
df_cp['Date'] = pd.to_datetime(df_cp['Date'])
```

```python
df_cp= df_cp.sort_values('Date')
```

```python
start_date = pd.to_datetime('2020-10-01')
```

```python
back_Date = start_date - pd.DateOffset(days=150)
back_Date
```

```python
df_cp = df_cp[(df_cp['Date'] >= back_Date) & (df_cp['Date'] < start_date)]
```

```python
df_cp[' adani Performce in %'] = (df_cp['Close_adani'] - df_cp['Open_adani'])/100
df_cp[' axis Performce in %'] = (df_cp['Close_axis'] - df_cp['Open_axis'])/100
df_cp[' asian Performce in %'] = (df_cp['Close_asian'] - df_cp['Open_asian'])/100
```

```python
df_best = df_cp.groupby('Symbol')['axis Performce in %'].mean().reset_index()
```
